```python
import pandas as pd

df = pd.read_csv('datasets/presidents.csv', index_col=0)
df.head()
```

```python
df['First'] = df['President']
df['Last'] = df['President']
df['First'] = df['First'].replace(" .+", "", regex=True)
df['Last'] = df['Last'].replace(".+ ", "", regex=True)
df.head()
```

```python
# del(df['First'])
# del(df['Last'])

def split(row):
    list = row['President'].split(' ')
    row['First'] = list[0]
    row['Last'] = list[-1]
    return row

df = df.apply(split, axis='columns')
df.head()
```

```python
del(df['First'])
del(df['Last'])
```

```python
pattern = "(?P<First>[\w]+)(?:.* )(?P<Last>[\w]+$)"
names = df['President'].str.extract(pattern)
for col in names.columns:
    df[col] = names[col]
df.head()
```

```python
df['Born'] = df['Born'].str.extract('(\w{3} \d{1,2}, \d{4})')
df['Born'].head()
```

```python
df['Born'] = pd.to_datetime(df['Born'])
# df['Born_1'] = pd.to_datetime(df['First'])
df.head(100)
```

```python
df['years_start'] = df['Age atstart of presidency'].str.extract('(\d+)(?:\s+years.+)')
df['days_start'] = df['Age atstart of presidency'].str.extract('(?:\d+\s+years,\s+)(\d+)(?:\s+days.+)')

df['years_end'] = df['Age atend of presidency'].str.extract('(\d+)(?:\s+years.+)')
df['days_end'] = df['Age atend of presidency'].str.extract('(?:\d+\s+years,\s+)(\d+)(?:\s+days.+)')

df['date_start'] = df['Age atstart of presidency'].str.extract('(?:.+?days)(.+)')
df['date_start'] = pd.to_datetime(df['date_start'])

df['date_end'] = df['Age atend of presidency'].str.extract('(?:.+?days)(\w{3} \d{1,2}, \d{4})')
df['date_end'] = pd.to_datetime(df['date_end'])

def delta(row):
    row['days_presidency'] = (row['date_end'] - row['date_start']).days
    return row
df = df.apply(delta, axis='columns')
df.head(100)

import numpy as np

print(np.mean(df['days_presidency'])/365)
print(np.min(df['date_start']))
print(np.max(df['date_end']))

```
